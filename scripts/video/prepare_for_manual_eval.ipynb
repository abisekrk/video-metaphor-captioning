```python
json_path = "/research/video_metaphor/benchmarks/Video-ChatGPT/results/3750_step_results.json"
file = open(json_path, "r")
chatgpt_results = json.load(file)
print(len(chatgpt_results))


true_captions = [ q["answer"] for q in chatgpt_results]
pred_captions = [ q["pred"].split("\n")[0] for q in chatgpt_results]
video_ids = [ q["id"] for q in chatgpt_results]
video_chatgpt_df = pd.DataFrame({"video_id": video_ids, "true_caption":true_captions, "pred_captions": pred_captions})
video_chatgpt_df = video_chatgpt_df.drop_duplicates(["video_id"], keep="last")
video_chatgpt_df.rename(columns={'pred_captions': 'videochatgpt'}, inplace=True)
video_chatgpt_df = video_chatgpt_df[["video_id", "videochatgpt"]]
print(len(video_chatgpt_df))
video_chatgpt_df
```

```python
json_path = "/research/video_metaphor/benchmarks/Video-LLaVA/results/best_output.json"
file = open(json_path, "r")
videollava_results = json.load(file)
print(len(videollava_results))
videollava_results[0]


true_captions = [ q["answer"] for q in videollava_results]
pred_captions = [ q["pred"].split("\n")[0] for q in videollava_results]
video_ids = [ q["id"] for q in videollava_results]
video_llava_df = pd.DataFrame({"video_id": video_ids, "true_caption":true_captions, "pred_captions": pred_captions})

video_llava_df = video_llava_df.drop_duplicates(["video_id"], keep="last")
video_llava_df.rename(columns={'pred_captions': 'video_llava'}, inplace=True)
video_llava_df = video_llava_df[["video_id", "video_llava"]]
print(len(video_llava_df))
video_llava_df
```

```python
# {
#     "id": 1,
#     "data": {
#         "text": "choose the best on fluency",
#         "m1": "M1: He is as tall as a elephant",
#         "m2": "M2: He is strong as a lion",
#         "m3": "M3: He is strong as a lion",
#         "m4": "M4: He is strong as a lion",
#         "m5": "M5: He is strong as a lion",
#        "video": "/data/local-files/?d=videos/sample.mp4"
#      }
#    }
def prepare_annotation_data(id_header, dataframe, save_fname):
    count = 1
    # base_path = "/data/local-files/?d=videos/"
    mapper_basepath = "/research/video_metaphor/LLaVA/playground/data/video_data/manual_eval_june/permutation_mapping/"
    dataset = []
    for index, row in dataframe.iterrows():
        row_dict = {}
        row_dict["id"] = id_header + str(count)
        data_dict = {}
        # data_dict["video"] = base_path + row["video_id"] + ".mp4"
        data_dict["video"] = f"<embed src='https://www.youtube.com/embed/{row['video_id']}'></embed>"
        with open(f"{mapper_basepath}{count}.json", 'r') as file:
            column_mapper = json.load(file)
        
        for key in column_mapper.keys():
            target_key = column_mapper[key]
            data_dict[target_key] = f"{target_key}: {row[key]}"
        row_dict["data"] = data_dict
        dataset.append(row_dict)
        count += 1

    with open(save_fname, 'w') as fp:
            json.dump(dataset, fp)
```

```python
prepare_annotation_data("ann_1_",an1_final,"/research/video_metaphor/LLaVA/playground/data/video_data/manual_eval_june/annotator1.json")
prepare_annotation_data("ann_2_",an2_final,"/research/video_metaphor/LLaVA/playground/data/video_data/manual_eval_june/annotator2.json")
prepare_annotation_data("ann_3_",an3_final,"/research/video_metaphor/LLaVA/playground/data/video_data/manual_eval_june/annotator3.json")
```
